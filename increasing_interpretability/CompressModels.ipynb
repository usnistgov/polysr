# Compress models

```python
import lzma
import shutil
from pathlib import Path
```

## Compress models

```python
files = list(Path("models").glob("*.pkl"))

for m in files:
    with open(m, "rb") as f_in:
        with lzma.open(str(m) + ".xz", "wb", preset=9) as f_out:
            shutil.copyfileobj(f_in, f_out)
```
