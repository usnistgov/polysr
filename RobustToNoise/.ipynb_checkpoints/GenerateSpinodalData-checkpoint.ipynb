# Generate data to predict the spinodal for Flory Huggins Theory

$\chi = \frac{1}{2} \left( \frac{1}{\phi N} + \frac{1}{1-\phi} \right)$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phi_lin = np.linspace(0.025, 0.975, 39)
logN_lin = np.logspace(0, 2, 17)

phi, N = np.meshgrid(phi_lin, logN_lin)

chi = 0.5 * (1 / (phi * N) + 1 / (1 - phi))
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

plt.rcParams["font.size"] = 12
plt.rcParams["axes.labelsize"] = 14
colors = plt.cm.plasma(np.linspace(0, 1, 6))
colors = sns.color_palette("plasma", 3)

plt.plot(phi[0, :], chi[0, :], "o", label=r"$N=1$", color=colors[0])
plt.plot(phi[8, :], chi[8, :], "o", label=r"$N=10$", color=colors[1])
plt.plot(phi[16, :], chi[16, :], "o", label=r"$N=100$", color=colors[2])
# plt.plot(np.transpose(phi), np.transpose(chi),'.-')
plt.xlabel(r"$\phi$")
plt.ylabel(r"$\chi_{spin}$")
# ax.legend()


plt.tight_layout()
plt.savefig("figures/SpinData.jpg", dpi=600)
```

```python
np.savetxt(
    "spin.csv",
    np.transpose(
        (np.ndarray.flatten(phi), np.ndarray.flatten(N), np.ndarray.flatten(chi))
    ),
    delimiter=",",
)
```
