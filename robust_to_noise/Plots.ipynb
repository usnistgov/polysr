# Create plots from the summarized data

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Error in y

## Compare the equation with complexity of 11

```python
data = np.genfromtxt("results/dy_eq11.csv", delimiter=",")

fig, ax = plt.subplots(figsize=(4, 3), sharey=True)

x_labels = [
    r"None",
    r"$10^{-3}$",
    r"$10^{-2}$",
    r"$10^{-1}$",
    r"$10^{0}$",
]
y_labels = data[1:, 0].astype(int)

sns.heatmap(
    data[1:, 1:] * 100.0,
    linewidth=0.3,
    annot=True,
    fmt=".0f",
    cmap="flare_r",
    annot_kws={"size": 11},
    # cbar_kws={"label": "Percent successful", "size": 13},
)
cbar = ax.collections[0].colorbar
cbar.set_label(label="Percent successful", size=13)
cbar.ax.tick_params(labelsize=11)

ax.set_xticklabels(x_labels, rotation=0, fontsize=11, horizontalalignment="center")
ax.set_yticklabels(y_labels, rotation=360, fontsize=11, horizontalalignment="right")
ax.set_ylabel("Number of datapoints", fontsize=13)
ax.set_xlabel("Noise", fontsize=13)

fig.text(0,.95,"a)",fontsize=12)
plt.tight_layout()
plt.savefig("figures/dy_eq11.jpg", dpi=600)
```
