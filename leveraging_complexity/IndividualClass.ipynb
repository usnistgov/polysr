# Individual Class

This notebook plots the results of running symbolic regression on each class individually.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Constant penalty = 2

```python
classlist = []
for i in range(6):
    classlist.append(pd.read_csv("symreg/eqns_class" + str(i + 1) + "_constpen.csv"))
```

```python
for i in range(6):
    print("Class", i)
    for j in range(len(classlist[i])):
        if classlist[i].iloc[j]["complexity"] == 9:
            print(
                j,
                classlist[i].iloc[j]["complexity"],
                classlist[i].iloc[j]["loss"],
                classlist[i].iloc[j]["equation"],
            )
```

```python
fig = plt.figure(figsize=(8, 4.8))
gs = fig.add_gridspec(2, 3, hspace=0, wspace=0)
(ax1, ax2, ax3), (ax4, ax5, ax6) = gs.subplots(sharex="col", sharey="row")

legendfont = 12

c = 0
s = 5
ax1.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="ABS, lit.",
)
ax1.plot(
    classlist[c].iloc[s]["complexity"],
    classlist[c].iloc[s]["loss"],
    marker="s",
    color=plt.cm.tab10(int(c)),
)

c = 1
s = 5
ax2.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="ABS",
)
ax2.plot(
    classlist[c].iloc[s]["complexity"],
    classlist[c].iloc[s]["loss"],
    marker="s",
    color=plt.cm.tab10(int(c)),
)

c = 2
s = 5
ax3.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="polycarbonate",
)
ax3.plot(
    classlist[c].iloc[s]["complexity"],
    classlist[c].iloc[s]["loss"],
    marker="s",
    color=plt.cm.tab10(int(c)),
)


c = 3
s = 4
ax4.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="d-PS",
)
ax4.plot(
    classlist[c].iloc[s]["complexity"],
    classlist[c].iloc[s]["loss"],
    marker="s",
    color=plt.cm.tab10(int(c)),
)

c = 4
s = 5
ax5.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="PS, 393 kDa",
)
ax5.plot(
    classlist[c].iloc[s]["complexity"],
    classlist[c].iloc[s]["loss"],
    marker="s",
    color=plt.cm.tab10(int(c)),
)

c = 5
ax6.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c + 1)),
    label="PS, 1571 kDa",
)

for ax in fig.get_axes():
    ax.label_outer()
    ax.set_yscale("log")
    ax.legend(loc="upper right", fontsize=legendfont)
    ax.set_ylim(10**-7, 10**4)
    ax.set_xticks(np.linspace(0, 25, 6))

ax4.set_xlabel(r"Complexity", fontsize=legendfont + 2)
ax5.set_xlabel(r"Complexity", fontsize=legendfont + 2)
ax6.set_xlabel(r"Complexity", fontsize=legendfont + 2)

ax1.set_ylabel(r"Mean squared error", fontsize=legendfont + 2)
ax4.set_ylabel(r"Mean squared error", fontsize=legendfont + 2)

ax4.tick_params(axis="x", labelsize=12)
ax5.tick_params(axis="x", labelsize=12)
ax6.tick_params(axis="x", labelsize=12)

ax1.tick_params(axis="y", labelsize=12)
ax4.tick_params(axis="y", labelsize=12)

plt.tight_layout()
plt.savefig("figures/IndivLossComplexConstPen.jpg", dpi=600)
```

WLF equation is found for all cases except for PS, 1571 kDa. It is not always clear it is the best equation, though. WLF equation is denoted as a square.

## Constant penalty = 1

```python
classlist = []
for i in range(6):
    classlist.append(pd.read_csv("symreg/eqns_class" + str(i + 1) + ".csv"))
```

```python
for i in range(6):
    print("Class", i)
    for j in range(len(classlist[i])):
        if classlist[i].iloc[j]["complexity"] == 7:
            print(
                j,
                classlist[i].iloc[j]["complexity"],
                classlist[i].iloc[j]["loss"],
                classlist[i].iloc[j]["equation"],
            )
```

```python
fig = plt.figure(figsize=(8, 4.8))
gs = fig.add_gridspec(2, 3, hspace=0, wspace=0)
(ax1, ax2, ax3), (ax4, ax5, ax6) = gs.subplots(sharex="col", sharey="row")

legendfont = 12

c = 0
ax1.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="ABS, lit.",
)

c = 1
ax2.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="ABS",
)

c = 2
ax3.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="polycarbonate",
)

c = 3
ax4.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="d-PS",
)

c = 4
ax5.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c)),
    label="PS, 393 kDa",
)

c = 5
ax6.plot(
    classlist[c]["complexity"],
    classlist[c]["loss"],
    marker="o",
    color=plt.cm.tab10(int(c + 1)),
    label="PS, 1571 kDa",
)


for ax in fig.get_axes():
    ax.label_outer()
    ax.set_yscale("log")
    ax.legend(loc="upper right", fontsize=legendfont)
    ax.set_ylim(10**-7, 10**4)
    ax.set_xticks(np.linspace(0, 25, 6))

ax4.set_xlabel(r"Complexity", fontsize=legendfont + 2)
ax5.set_xlabel(r"Complexity", fontsize=legendfont + 2)
ax6.set_xlabel(r"Complexity", fontsize=legendfont + 2)

ax1.set_ylabel(r"Mean squared error", fontsize=legendfont + 2)
ax4.set_ylabel(r"Mean squared error", fontsize=legendfont + 2)

ax4.tick_params(axis="x", labelsize=12)
ax5.tick_params(axis="x", labelsize=12)
ax6.tick_params(axis="x", labelsize=12)

ax1.tick_params(axis="y", labelsize=12)
ax4.tick_params(axis="y", labelsize=12)

plt.tight_layout()
plt.savefig("figures/IndivLossComplex.jpg", dpi=600)
```

The WLF equation is never found. Instead a different equation of the form $(c1 / (c2 + dT)) + c3$ is found instead. This has three constants and in practice is likely less desired compared to WLF.
