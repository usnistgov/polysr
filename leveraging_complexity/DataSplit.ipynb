# Create folds based on the data

This notebook performs a stratified 4 fold split based on the class of each data point. Original data is stored in `data/WLF_data.csv`. Resulting data is saved in `data/WLF_data_folds.csv`. Please note that the splits used in the publication are already saved in `data`, so running this script will overwrite them.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv("data/WLF_data.csv")
df.head()
```

```python
sfolds = StratifiedKFold(n_splits=4, shuffle=True)

x_in = df.iloc[:, 1:4].to_numpy()
y_in = df.iloc[:, 4].to_numpy()

T = (x_in[:, 0]).reshape(-1, 1)
groups = x_in[:, 2].astype(int)
```

```python
sfolds.get_n_splits(T, groups)

testfold = np.ones(len(T)) * 4

for i, (train_index, test_index) in enumerate(sfolds.split(T, groups)):

    testfold[test_index] = np.zeros(len(test_index))+i
```

```python
df["fold"] = testfold.astype(int)
```

```python
df.to_csv('data/WLF_data_folds.csv', index=False)
```
